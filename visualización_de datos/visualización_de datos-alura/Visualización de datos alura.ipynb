# Guía de Visualización de Datos en Python con Matplotlib, Seaborn y Plotly

Esta guía tiene como objetivo ayudarte a comprender cómo visualizar datos de interés utilizando herramientas de visualización en Python como **Matplotlib**, **Seaborn** y **Plotly**. A lo largo del documento, aprenderás a:

- Representar datos en diferentes tipos de gráficas.  
- Personalizar y mejorar la apariencia de las visualizaciones.  
- Exportar las gráficas para incluirlas en reportes o presentaciones.  
- Crear gráficas interactivas, especialmente gráficos de líneas, que permiten explorar la información de manera dinámica.


```python
#Para comenzar subimos las librerias que usaremos para tratar los datos y luego visualizarlos.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#vamos a estraer los datos de un documento csv y visualizar el contenido
df = pd.read_csv('/content/drive/MyDrive/visualización_de datos-alura/inmigrantes_canada.csv')
df.head()
```

```python
#Con shape podemos saber de cuantas columnas y filas esta conformada la data
df.shape
```

```python
#Estamos imprimiendo la columna de Regiones, para saber cuales tiene dentro.
print(df['Region'].unique())

```

```python
#pedimos la informacion de nuestras columnas en cuento a que tipo de datos tenemos
df.info()
```

```python
#Convierte la columna 'Pais' del DataFrame df en el índice del DataFrame.
df.set_index('Pais',inplace=True)
df
```

```python
#variable
años = list(map(str,range(1980,2014)))
años
```

```python
colombia = df.loc['Colombia',años]
colombia
```

```python
#data frame con diccionario
col_dict = {'Año':colombia.index.tolist(),
            'Inmigrantes':colombia.values.tolist()}
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
#primeras graficas:identificar patron de tendencia
plt.figure(figsize = (8,4))#modificamos el tamaño
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010']) #solo visualizaremos esos años
plt.show()
```

```python
plt.figure(figsize = (8,4))#modificamos el tamaño
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010']) #solo visualizaremos esos años
plt.title('Inmigración de Colombia hacia Canada')# titulo general
plt.xlabel('Año')#titulo eje x
plt.ylabel('Numero de inmigrantes')#titulo eje y
plt.show()
```

```python
#generamos variables para tener nuestro lienzo y grafica
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))#este nos permite que automaticamente busque un rango siguiendo el intervalo de 5 a 5
ax.set_title('Inmigración de colombianos hacia Canada\n1980 - 2013') #titulo principal
ax.set_xlabel('Año')#titulo eje x
ax.set_ylabel('Numero de Inmigrantes')#titulo eje y
plt.show()
```

```python
#grafico de caja
fig, axs = plt.subplots(1,2,figsize=(10,3)) #aqui declaramos que queremos 2 figuras
axs[0].plot(datos_col['Año'],datos_col['Inmigrantes'])
axs[0].set_title('Inmigración de colombianos hacia Canada?1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Numero de inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()#nos crea cuadricula

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxtplot de la inmigración de colombianos\nhacia Canada 1980 - 2013')
axs[1].set_xlabel('Colomnia')
axs[1].set_ylabel('Numero de inmigrantes')
axs[1].grid()
plt.show()
```

```python
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)#espacios entre graficas
fig.suptitle('Inmigranción de los 4 mayores paises sudamericanos\nhacia Canada desde  1980 - 2013')#titulo general para todo el recuadro d graficas

axs[0,0].plot(df.loc['Colombia',años])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil',años])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina',años])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú',años])
axs[1,1].set_title('Perú')
# en esta parte ponemos un rango max y min para y en todas las graficas
ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)


for ax in axs.flat:#en esta parte configuramos los titulos y el rangoautomatico
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Numero de inmigrantes')
  ax.grid()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,3))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'],lw=1,color='g', marker='o')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))#este nos permite que automaticamente busque un rango siguiendo el intervalo de 5 a 5
ax.set_title('Inmigración de colombianos hacia Canada\n1980 - 2013', fontsize=18, loc ='left') #titulo principal, aqui tambien al final podemos cambiar la fuente
ax.set_xlabel('Año', fontsize=14)#titulo eje x
ax.set_ylabel('Numero de Inmigrantes')#titulo eje y
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)
ax.grid(linestyle='--')
plt.show()
```

```python
sudamerica = df.query('Region == "América del Sur"')
sudamerica.head(10)
```

```python
#grafica de totales de migrantes de estos paises
colores = ['royalblue', 'orange', 'forestgreen', 'orchid','purple',
           'brown','slateblue','gray', 'olive','navy', 'teal','tomato']

fig, ax =plt.subplots(figsize=(12,5))

ax.bar(sudamerica.index, sudamerica['Total'] , color=colores)
ax.set_title('Inmigracion de sudamericanos hacia Canada\ndurante el periodo de 1980 hasta 2013',
            loc='left',fontsize=18)

ax.set_xlabel('')
ax.set_ylabel('Numero de Inmigrantes',fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()

```

```python
colores = ['royalblue', 'orange', 'forestgreen', 'orchid','purple',
           'brown','slateblue','gray', 'olive','navy', 'teal','tomato']

fig, ax =plt.subplots(figsize=(10,3))

ax.barh(sudamerica.index, sudamerica['Total'] , color=colores)
ax.set_title('Inmigracion de sudamericanos hacia Canada\ndurante el periodo de 1980 hasta 2013',
            loc='left',fontsize=18)

ax.set_ylabel('')
ax.set_xlabel('Numero de Inmigrantes',fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
sudamerica_sorted = sudamerica.sort_values('Total',ascending=True)
```

```python
colores = ['royalblue', 'orange', 'forestgreen', 'orchid','purple',
           'brown','slateblue','gray', 'olive','navy', 'teal','tomato']

fig, ax =plt.subplots(figsize=(10,3))

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'] , color=colores)
ax.set_title('Inmigracion de sudamericanos hacia Canada\ndurante el periodo de 1980 hasta 2013',
            loc='left',fontsize=18)

ax.set_ylabel('')
ax.set_xlabel('Numero de Inmigrantes',fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
colores =[]
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
    colores.append('silver')

fig, ax =plt.subplots(figsize=(10,3))

#quitamos lo que no queremos ver con es decir los numeros al rededor x y y
ax.set_frame_on(False)
ax.xaxis.set_visible(False)
ax.tick_params(axis='y',which='both',size=0)

#barras horizontales
ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'] , color=colores)
ax.set_title('Sudamérica: Colombia es el segundo país con más migrantes\nhacia Canada durante el periodo de 1980 hasta 2013',
            loc='left',fontsize=16)

ax.set_ylabel('')
ax.set_xlabel('Numero de Inmigrantes',fontsize=14)
#ax.xaxis.set_tick_params(labelsize=12)#los comentamos porque no necesitamos ya que quitamos esa info
#ax.yaxis.set_tick_params(labelsize=12)

for i,j in enumerate(sudamerica_sorted['Total']):
 ax.text(j+20,i,str(j),color='black',fontsize=8.5,ha='left',va='center')

fig.savefig('/content/drive/MyDrive/visualización_de datos-alura/inmigrantes_sudamericanos_canada.png',
            transparent=False,dpi=300,bbox_inches='tight')



plt.show()
```

```python
#modificamos estaparte para que aparesca la paleta de colores
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax = sns.barplot(data=top_10,x='Total',y=top_10.index,orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 paises con mayor migración hacia canada\n1980 - 2013',
            loc='left',fontsize=16 )
  ax.set_xlabel('Numero de inmigrantes',fontsize=14)
  ax.set_ylabel('')

  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black',fontsize=9,ha='left',va='center')
  plt.show()
```

```python
generar_grafico('Blues_r')
```

```python
generar_grafico('rocket')
```

```python
#modificamos para que funcione con siborn
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set(xticklabels=[])
  ax = sns.barplot(data=top_10,x='Total',y=top_10.index,orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 paises con mayor migración hacia canada\n1980 - 2013',
            loc='left',fontsize=16 )
  ax.set_xlabel('Numero de inmigrantes',fontsize=14)
  ax.set_ylabel('')
  sns.despine()
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black',fontsize=9,ha='left',va='center')
  plt.show()
```

```python
sns.set_theme(style='darkgrid')
generar_grafico('tab10')
```

```python
#cambiamos a otro estilo
sns.set_theme(style='whitegrid')
generar_grafico('tab10')
```

```python
sns.set_theme(style='white')
generar_grafico('tab10')
```

```python
fig = px.line(datos_col,x='Año',y='Inmigrantes',
              title='Inmigrantes de colombianos hacia canada ')
fig.update_traces(line_color='green',line_width=3)
fig.update_layout(width=600,height=400,xaxis={'tickangle':-45},
                  xaxis_title='Año', yaxis_title='Numero de inmigrantes',font_family='Arial',font_size=14, font_color='grey')
fig.show()
```
